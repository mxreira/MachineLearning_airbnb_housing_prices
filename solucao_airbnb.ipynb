### Funções criadas para a análise de outliers

```python

def limites(coluna: pd.Series):

    q1 = coluna.quantile(0.25)
    q3 = coluna.quantile(0.75)

    amplitude = q3 - q1

    limite_inferior = q1 - (1.5 * amplitude)

    limite_superior = q3 + (1.5 * amplitude)

    return limite_inferior, limite_superior

def excluir_outliers(df, nome_coluna):

    qtde_linhas_inicial = df.shape[0]

    limite_inferior, limite_superior = limites(df[nome_coluna])

    df = df.loc[(limite_inferior <= df[nome_coluna]) & (df[nome_coluna] <= limite_superior), :]

    qtd_linhas_novas = df.shape[0]

    linhas_removidas = qtde_linhas_inicial - qtd_linhas_novas

    return df, linhas_removidas


    
```

```python
def diagrama_caixa(coluna):
    fig, (ax1, ax2) = plt.subplots(1, 2)

    fig.set_size_inches(15, 5)

    sns.boxplot(x=coluna, ax= ax1)

    ax2.set_xlim(limites(coluna))

    sns.boxplot(x=coluna, ax= ax2)

def histograma(coluna):

    plt.figure(figsize= (15,5))

    sns.histplot(coluna)

def grafico_barra(coluna):

    plt.figure(figsize= (15,5))


    ax = sns.barplot(x= coluna.value_counts().index, y= coluna.value_counts())

    ax.set_xlim(limites(coluna))
```

## Encoding

Precisamos ajustar as features para facilitar o trabalho do modelo futuro (features de categoria, true e false, etc)

- Features de True/False: 1 e 0 respectivamente.

- Features de Categoria (que são textos) que não tem ordem: método encoding de Variáveis Dummies.

### - True/False

```python
colunas_tf = ['host_is_superhost', 'instant_bookable', 'is_business_travel_ready']

df_airbnb_cod = df_airbnb.copy()

for coluna in colunas_tf:
    
    df_airbnb_cod.loc[df_airbnb_cod[coluna] == 't', coluna] = 1
    df_airbnb_cod.loc[df_airbnb_cod[coluna] == 'f', coluna] = 0

print(df_airbnb_cod.iloc[0])
```

<h2>Modelo de Previsão</h2>

- Serão selecionadas as métricas de R² e RSME

```python
def avaliar_modelo(nome_modelo: str, y_teste, previsao):

    r2 = r2_score(y_teste, previsao)
    RSME = np.sqrt(mean_squared_error(y_teste, previsao))

    return 'Modelo {}: \nR²: {:.2%}\nRSME: {:.2f}'.format(nome_modelo, r2, RSME)
```

- Escolha dos Modelos a Serem Testados
    1. RandomForest
    2. LinearRegression
    3. Extra Tree

```python
modelo_rf = RandomForestRegressor()
modelo_lr = LinearRegression()
modelo_et = ExtraTreesRegressor()

modelos = {'RandomForest': modelo_rf,
           'LinearRegression': modelo_lr,
           'ExtraTrees': modelo_et}

y = df_airbnb_cod['price']
X = df_airbnb_cod.drop('price', axis= 1)
```

- Separação das Variáveis em treino e teste + Treino do Modelo

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state= 42) # random_state serve para padronizar a separação dos dados de treino e teste

for nome_modelo, modelo in modelos.items():
    start_time = time.time()
    #treinar
    modelo.fit(X_train, y_train)

    #testar
    previsao = modelo.predict(X_test)
    print(avaliar_modelo(nome_modelo, y_test, previsao))
    end_time = time.time()

    duracao = end_time - start_time

    print(f'Duração: {duracao:.2f} segundos')
```

<h2>Análise do Melhor Modelo</h2>

```python
for nome_modelo, modelo in modelos.items():
    start_time = time.time()

    #testar
    previsao = modelo.predict(X_test)
    print(avaliar_modelo(nome_modelo, y_test, previsao))
    end_time = time.time()

    duracao = end_time - start_time

    print(f'Duração: {duracao:.2f} segundos')
```
